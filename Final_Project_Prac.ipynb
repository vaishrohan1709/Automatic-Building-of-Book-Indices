```python
def getFrequencyFromNgrams( filename, n ):
    file1 = open(filename)
    finalText = file1.read()
    text = nltk.word_tokenize(finalText)
    list2 = nltk.pos_tag(text)
    list1 = []
    for a in list2:
        list1.append(a[0].lower())
    word_freq = {}
    for word in sorted(list1):
        if word in word_freq:
            word_freq[word]+=1
        else:
            word_freq[word] = 1
    
    GNgram_Freq_word = {}
    GNgram_Freq_volume = {}
    distict_list = list(set(list1))

    print(len(distict_list))
    i = 0
    for l in distict_list:
        for d in dataframes:
            try:
                GNgram_Freq_word[l] = d.loc[l].count_w
                GNgram_Freq_volume[l] = d.loc[l].count_v
                break
            except:
                continue
    
    #df = pd.DataFrame.from_dict(GNgram_Freq_word,index=['i',])
    df = pd.DataFrame.from_dict(GNgram_Freq_word, orient='index')
    df1 = pd.DataFrame.from_dict(GNgram_Freq_volume, orient='index')
    df3 = pd.DataFrame.from_dict(word_freq, orient='index')
    
    df2 = df.join(df1, how='left', lsuffix='_word_ngram_freq', rsuffix='_volume_ngram_freq' )
    df4 = df2.join(df3, how='left', lsuffix='ngram_freq', rsuffix='_freq' )
    
    df4.rename( columns={0: 'freq', '0_volume_ngram_freq': 'volume_ngram_freq', '0_word_ngram_freq': 'word_ngram_freq' }, 
                inplace=True)
    
    print(df4.head(3))
    
    new_file = 'word_freq_gngramfreq_volfreq/' + filename.split('/')[1].split('.')[0] + '.csv' 
    df4.to_csv(new_file)
```

```python
import numpy as np
def normalize(filename):
    df = pd.read_csv(filename)
    
    print(df.head(2))
    
    mean = np.mean(df['word_ngram_freq'])
    std = np.std(df['word_ngram_freq'])
    df['word_ngram_freq'] = (df['word_ngram_freq']-mean)/std
    
    m = np.min(df['word_ngram_freq'])
    df['word_ngram_freq'] = df['word_ngram_freq'] - m
    
    mean = np.mean(df['volume_ngram_freq'])
    std = np.std(df['volume_ngram_freq'])
    df['volume_ngram_freq'] = (df['volume_ngram_freq']-mean)/std
    
    m = np.min(df['volume_ngram_freq'])
    df['volume_ngram_freq'] = df['volume_ngram_freq'] - m
    
    mean = np.mean(df['freq'])
    std = np.std(df['freq'])
    df['freq'] = (df['freq']-mean)/std
    
    m = np.min(df['freq'])
    df['freq'] = df['freq'] - m
    
    df['score'] = (1/df['word_ngram_freq']) * df['freq'] * df['volume_ngram_freq']
    new_file = 'ngram_score/' + filename.split('/')[1].split('.')[0] + '.csv'
    df.to_csv(new_file)
```

```python
normalize('word_freq_gngramfreq_volfreq/1.csv')
normalize('word_freq_gngramfreq_volfreq/2.csv')
normalize('word_freq_gngramfreq_volfreq/3.csv')
normalize('word_freq_gngramfreq_volfreq/4.csv')
normalize('word_freq_gngramfreq_volfreq/5.csv')
normalize('word_freq_gngramfreq_volfreq/6.csv')
normalize('word_freq_gngramfreq_volfreq/7.csv')
normalize('word_freq_gngramfreq_volfreq/8.csv')
normalize('word_freq_gngramfreq_volfreq/9.csv')
normalize('word_freq_gngramfreq_volfreq/10.csv')
normalize('word_freq_gngramfreq_volfreq/11.csv')
```

```python
from operator import itemgetter  

def removeTwoThreeNumberWords(filename):
    df = pd.read_csv(filename)
    df.rename( columns={'Unnamed: 0.1': 'word'}, inplace=True)
    df.drop('Unnamed: 0', axis=1, inplace=True)
    #print(df.head(3))
    #i=0
    word_score = {}
    for index, row in df.iterrows():
        if len(str(row['word'])) > 2 and row['score'] > 0:
            word_score[str(row['word'])] = row['score']
            
    dc_sort = sorted(word_score.items(),key = itemgetter(1),reverse = True)
    
    new_file = 'gngram_index/' + filename.split('/')[1].split('.')[0] + '.txt'
    
    appendFile = open(new_file,'a')
    
    for r in dc_sort:
        appendFile.write(r[0]+'\n')
    appendFile.close
```

```python
def evaluate( test, actual ):
    with open(actual) as fin:
        d = {}
        i_list=[]
        for line in fin:
            i_list.append(line)

    #print(i_list)
    num_lines = sum(1 for line in open(test))
    num_lines_OF=sum(1 for line in open(actual))

    print("actual indices ", num_lines_OF)
    print("predicted indices ", num_lines)

    trials = 15
    num_indices_list = [10, 30, 50, 100,130,150,170,200, 220, 250, 300, 320, 350, 400, 420 ]
    mylist=[]
    mylist1=[]
    for i in range(0,trials):
        count=0
        num = num_indices_list[i]
        mylist1.append(i+1)
        with open(test) as fin1:
            for j in range(0, num):
                line=fin1.readline()
                if line in i_list :
                    count=count+1
                    
        #print('Number of matches:' , count)
        if num < num_lines_OF:
            percentage = int(count*100/num)
        else:
            percentage = int(count*100/num_lines_OF)
        mylist.append(percentage)
        if i>=5 :
            print('Number of matches:' , count)
            print('Percentage Matched: ', percentage)
            pyplot.figure(figsize=(20,10))
            pyplot.xlabel('Number of Trials')
            pyplot.ylabel('Percentage match')
            pyplot.bar(mylist1, mylist, width=0.3)
            pyplot.show()
```
